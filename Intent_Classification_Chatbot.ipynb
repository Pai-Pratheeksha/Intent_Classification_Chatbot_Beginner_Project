```python
import random

intents = [
    {"intent": "greeting", "text": "hello"},
    {"intent": "greeting", "text": "hi"},
    {"intent": "greeting", "text": "hey"},
    {"intent": "greeting", "text": "good morning"},
    {"intent": "greeting", "text": "good evening"},

    {"intent": "about", "text": "who are you"},
    {"intent": "about", "text": "what is your name"},
    {"intent": "about", "text": "what do you do"},

    {"intent": "cyber_tips", "text": "give me a cyber tip"},
    {"intent": "cyber_tips", "text": "how to stay safe online"},
    {"intent": "cyber_tips", "text": "internet safety tip"},

    {"intent": "goodbye", "text": "bye"},
    {"intent": "goodbye", "text": "good night"},
    {"intent": "goodbye", "text": "see you"},
]

responses = {
    "greeting": ["Hey there!", "Hello!", "Hi! How can I help?"],
    "about": ["I'm your Cyber Assistant Chatbot.", "Iâ€™m an AI chatbot built to help with cybersecurity."],
    "cyber_tips": [
        "Always use strong, unique passwords and enable 2FA.",
        "Donâ€™t click on suspicious links or attachments.",
        "Keep your antivirus updated."
    ],
    "goodbye": ["Bye! Stay safe online.", "Goodbye!", "Take care and stay secure!"]
}
```

```python
texts = [item['text'] for item in intents]
labels = [item['intent'] for item in intents]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(texts)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X, labels)
```

```python
def chatbot_response(user_input):
    X_test = vectorizer.transform([user_input])
    predicted_intent = model.predict(X_test)[0]

    bot_reply = random.choice(responses[predicted_intent])
    return bot_reply
```

```python
print("ðŸ¤– Cyber Assistant: Hello! Type 'quit' to exit.\n")

while True:
    user_input = input("You: ")
    if user_input.lower() == "quit":
        print("ðŸ¤– Cyber Assistant: Goodbye!")
        break

    reply = chatbot_response(user_input)
    print("ðŸ¤– Cyber Assistant:", reply)
```
